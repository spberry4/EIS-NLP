## Importing Libraries needed for wordcloud

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
text = '/content/drive/MyDrive/Data/final_eisdata_corpus.txt'
```

```python
text_file = open(text)
corpus = text_file.read()
```

## Creating the Wordcloud and displaying the generated image

```python
# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480).generate(corpus)
 
# Display the generated image:
plt.figure(figsize=(15,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
